### Importing the packages

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

### Loading the dataset

```python
titanic_train = pd.read_csv('train.csv')
titanic_train.head()
```

### Checking for null values

```python
titanic_train.isna().sum()
```

### Converting String to Int

```python
l_enc = preprocessing.LabelEncoder()
titanic_train['Sex'] = l_enc.fit_transform(titanic_train['Sex'])
titanic_train['Embarked'] = l_enc.fit_transform(titanic_train['Embarked'])
```

```python
titanic_train.head(20)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 1000,max_features=2,oob_score=True)
features = ['Sex','Pclass','SibSp','Age','Parch','Fare','Embarked']
rf_model.fit(X=titanic_train[features],y=titanic_train['Survived'])
rf_model.oob_score_
```

```python
for features,imp in zip(features,rf_model.feature_importances_):
    print(features,imp)
```

#### Inference : This shows that the important features are Sex, Age and Fare.

### Decision Tree

```python
tree_model = tree.DecisionTreeClassifier(max_depth=8)
predictors = pd.DataFrame([titanic_train['Sex'],titanic_train['Age'],titanic_train['Fare']]).T
tree_model.fit(X = predictors,y=titanic_train['Survived'])
```

```python
with open('Dtree1.dot','w') as f:
    f= tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f)
    
```

```python
tree_model.score(X=predictors,y=titanic_train['Survived'])
```

### Prediction

```python
titanic_test = pd.read_csv('test.csv')
titanic_test.head()
```

```python
titanic_test.isna().sum()
```

```python
titanic_test['Sex'] = l_enc.fit_transform(titanic_test['Sex'])
titanic_test.head()
```

```python
test_features = pd.DataFrame([titanic_test['Sex'],titanic_test['Age'],titanic_test['Fare']]).T
test_predict = tree_model.predict(X=test_features)
```

```python
predicted_op = pd.DataFrame({'PassengerId':titanic_test['PassengerId'],'Survived':test_predict})
```

```python
predicted_op.to_csv('Output.csv',index=False)
```
